The rocket will be flying in a straight line with constant speed. So we have the speed of the rocket, we have the asteroid's trajectory - this program will help us to find the exact time to launch the rocket.

```python
import matplotlib.pyplot as plt
import numpy as np
import math
g = 9.81
# the asteroid is falling straight downwards, we have just Vy = g.t
# we also know the time for the asteroid reaching the y_point where we want to destroy it
def destroy_asteroid(V_rocket,angle, Vo_asteroid, x_asteroid, y_asteroid):
    t = x_asteroid / (V_rocket * math.cos(math.radians(angle)))
    y1 = t * V_rocket * math.sin(math.radians(angle))
    
    distance = (y_asteroid - y1)
    t1 = (-Vo_asteroid + math.sqrt(Vo_asteroid ** 2 + 8 * g *distance)) / g
    
    
    ys = [y_asteroid - Vo_asteroid * t - 1/2 * g * t ** 2 for t in np.arange(0, t1)]
    xs = [x_asteroid] * len(ys) # asteroid trajectory
   
    
    ys1 = [V_rocket * t * V_rocket * math.sin(math.radians(angle)) for t in np.arange(0, t1)]
    xs1 = [V_rocket * t * V_rocket * math.cos(math.radians(angle)) for t in np.arange(0, t1)]
    
    plt.plot(xs, ys)
    plt.plot(xs1, ys1)
    plt.show()
    print("The coordinates of the shooted asteroid:")
    print(x_asteroid, y1)
    print("The time we have to launch the rocket before the asteroid reaches the stratosphere(or the point where we have just a few seconds to target it)")
    return t1 - t
   
    # V_rocket . sina .t + Vo_asteroid .t + 1/2 . g. t^2 = y we sum the two trajectories to find the 

    
```
