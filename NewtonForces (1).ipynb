### Physics Principles
So let's start with some basics
You can't haven't heard about the well-known Newtonian Principles, which lay on te basics of Mechanics:)
Let's remind them:)

First Law: Every particle continues in a state of rest or uniform in a straight line unless it is compelled to change 
that state by forces impressed on it. 

Second Law: The change of motion is proportional to the force impressed and is made in a direction of the straight line in which the force is impressed.

Third Law: If an object A exerts a force on object B, then object B must exert a force of equal magnitude and opposite direction back on object A.

Let's see how the second law works!
The sum of all vectors of the powers, which are impressed on the body, is equal to the mass of the object, multiplied with the vector of the acceleration.

```python
import numpy as np
import matplotlib.pyplot as plt

def vectorize_powers(vectors):
    
    
    plt.gca().set_aspect("equal")
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    
    for vector in vectors:
        start_x = vector[0]
        start_y = vector[1]
        
        len_x = vector[2] - vector[0]
        len_y = vector[3] - vector[1]
        
        plt.quiver(start_x, start_y, len_x, len_y,angles="xy", scale_units = "xy", scale = 1)
    plt.show()
    
    
    
    
    
    
```

```python
def draw_scale():
   
    
    plt.gca().set_aspect("equal")
    plt.xlim(-7,7)
    plt.ylim(-7,7)
    
    line = plt.Line2D((-7, 7), (-1, -1), lw=1.5)
    plt.gca().add_line(line)
    rectangle = plt.Rectangle((-1,-1), 2, 2, fc='black')
    plt.gca().add_patch(rectangle)

def draw_support_power(vector):
    draw_scale()
    m = 0.4 # the mass of the object
    g = 9.81
    plt.quiver(0, 0, vector[0], vector[1] ,angles="xy", scale_units = "xy", scale = 1, color="orange", label="Force of motion")
    
    plt.quiver(0, 0, 0, m*g - vector[1] ,angles="xy", scale_units = "xy", scale = 1, color="blue", label="Support force")
    plt.legend()
    plt.show()
    
    
    
```

Lets see what happens if we add acceleration to the equation:)
There are two types of movement-when a < 0 and when a > 0, in the first case the speed is decreasing, while in the other case- is increasing:
$$ V = Vo ± at $$
$$ S = Vot ± \frac{at^2}{2} $$ 

Let's make a graph:)

Now lets make a graph of the motion- we will prove that the greater the impressed power is, the bigger the acceleration is-the faster the object is moving.

```python
def graph_when_a_negative(Vo, a, time):
    times = [t for t in range(time)]
    S = [(Vo*t - 1/2 * a * t ** 2) for t in times]
    plt.plot(times, S)
    plt.xlabel("Time [s]")
    plt.ylabel("Distance [m]")
    plt.show

def graph_when_a_positive(Vo, a, time):
    times = [t for t in range(time)]
    S = [(Vo*t + 1/2 * a * t ** 2) for t in times]
    plt.plot(times, S)
    plt.xlabel("Time [s]")
    plt.ylabel("Distance [m]")
    plt.show
```

```python
graph_when_a_negative(10, 2, 10)
```

```python
graph_when_a_positive(10, 2, 10)
```

We can notice the difference between the two graphs, right!
In the first one, the object is decresing its speed until the speed reaches zero, then we can see how the distance becomes a negative number, while the second one is constantly increasing its speed.
Now we will increase time: The distance become a negative number,no way - that means that the object is in rest(it is not moving.)

```python
graph_when_a_negative(10, 2, 100)
```

Now with these formulas above, we can solve problems.
For example the most simple problem: we have two cars with speeds V1 and V2 and we want to find after how many hours they will meet.

```python
def time1(V1,V2,S):
    time = S/(V1+V2)
    return time

time1(10, 20, 900)
#if they are moving against each other
```

```python
def time2(V1,V2,S):
    time = S/abs(V1-V2)
    return time
time2(10, 20, 900)
#if they are moving in one direction
```

Now let me explain the Earth acceleration g. 
$$ G = mg $$
g = 9.81 m/s^2
Now we can state a real-life problem - falling objects,imagine that your neighbour from the sixth floor is cleaning her window and accidently pushes her flower pot, so you will be able to calculate how much time you have until you get your head hit by it.I know that calculating the time will be your most insignificant problem, but who knows:) 

So the formula is:
$$ h = Vot ± \frac{gt^2}{2} $$
The two cases - if we throw the object upward the sign in front of g will be (-)
              - if the object is just falling the sigh will be (+)
Let's return to the clumsy neighbour, the pot is just pushed, so its Vo will be zero, then we can use the formula:
$$ h = \frac{gt^2}{2} $$

The time for falling doesn't depend on the mass of the body, because the acceleration is a constant for all bodies.

```python
import math
def find_time_for_falling(h):
    g = 9.81
    time = math.sqrt(2*h / g)
    return time

find_time_for_falling(25)
```

```python
Oh no, you will have just 2.26 seconds to escape the situation without getting hurt.
```

```python
def height_time(h):
    g = 9.81
    heights = [h1 for h1 in range(h, 0, -1)]
    time = [math.sqrt(2*h / g) for h in reversed(heights)]
    plt.plot(time, heights)
    plt.ylabel("Height[m]")
    plt.xlabel("Time[seconds]")
    plt.show()

height_time(90)
```

Here you are some other functions for finding distances during acceleration movements.

```python
def find_a(Vo, S, t): # a > 0
    a = 2 * (S - Vo*t) / t ** 2
    return a

def find_a_prim(Vo, S, t):# a < 0
    a = -2 * (S - Vo*t) / t ** 2
    return a

find_a(100, 1000, 5)
```

```python
def find_Vo(a, t, S):
    if a > 0:
        Vo = (S - 1/2 * a * t **2) / t
    else:
        Vo = (S + 1/2 * a * t **2) / t
        
    return Vo

find_Vo(2, 10, 1000)
```
