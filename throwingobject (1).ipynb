So t = (2.Vo.sina / g )//the time the object for reaching the surface 


x(t) = Vo.cosa.t = Vo.cosa.2.sina / g = Vo^2.sin2a/g
y(t) = Vo.sina.t - 1/2.g.t^2 = Vo^2.sina^2/g



```python
import math

def find_distance(angle, Vo):
    Vox = Vo*math.cos(math.radians(angle))
    Voy = Vo*math.sin(math.radians(angle)) # we convert the degrees into radians
    
    g = 9.81
    time = (2*Voy/ g)
    S = Vox * time
    return S
```

```python
find_distance(45, 10)
```

```python

def find_height(angle, Vo):
    g = 9.81
    Vox = Vo*math.cos(math.radians(angle))
    Voy = Vo*math.sin(math.radians(angle))
    height = Voy**2 / g
    return height
    
    
```

We can make a graph of the trajectory of the object

```python
def graph_traectory(angle, Vo):
    g = 9.81
    Vox = Vo*math.cos(math.radians(angle))
    Voy = Vo*math.sin(math.radians(angle))
    
    time =  2 * int(Voy / g)
    #print(time)
    heights = [Voy*t-1/2 * g *t**2 for t in range(0, time + 1)]
    #print(heights)
    
    distances = [Vox*t for t in range(0, time + 1)]
    plt.plot(distances, heights)
    plt.xlabel("Distance [m]")
    plt.ylabel("Height [m]")
    plt.show()    
    
    
```

Now we can make a function for finding the perfect angle to shoot at exact coordinates, like in "Angry Birds".
We can use it also for basketball games - using the physical principles we can make the game look more realistic.

After solving one of the hardest equations that I have ever seen - a trigonometric one - we can find the angle.

```python
import numpy
def shoot_at_target(Vo, x, y): # we write the coordinates of the target - x, y;and Vo - starting velocity
    g = 9.81
    
    A = x / y
    B = x**2 * g / (2 * y * Vo**2)
    # I put cosa^2 = z, and A, B for coefficients and solve the quadric equation 
    
    z1 = (A**2 - 2*B + math.sqrt(A**4 + 4*B**2 - 2*A*B**2 - A**2*B**2 - B**2)) / 4 * (A**2 + 1) * B**2
    a = math.sqrt(z1)
    return numpy. rad2deg(math.acos(a))
```

Now let me challenge you - how are you gonna try to shoot on a moving target:)
For example a falling object - this function will help us to find the exact time to shoot, so to be able to hit the target.But we will need and the value of the angle this time

```python
import numpy
def shoot_at_falling_target(Vo, angle, x, h): # we write the coordinates of the target - x, y;and Vo - starting velocity
    g = 9.81
    time = x / Vo * math.cos(math.radians(angle)) # we find the time to reach the coordinate of x
    y = Vo * math.sin(math.radians(angle)) * time - 1/2 * g * time ** 2
    
    time1 = math.sqrt(2*(h-y) / g)
    
    
    return time1 - time # we find the exact time to shoot the arrow, after the object started falling
```

```python
shoot_at_falling_target(10, 30, 5, 10)
```

```python
import numpy as np
def graph_shooted_target(Vo, angle, h, x):
    g = 9.81
    time = math.sqrt(2*h/g)
    ys = [1/2 * g * t ** 2 for t in np.linspace(0, time)]
    xs = [x] * len(ys)
    
    
    Vox = Vo * math.cos(math.radians(angle))
    Voy = Vo * math.sin(math.radians(angle))
    time1 = x / Vox
    heights = [Voy * t - 1/2 * g * t**2 for t in np.linspace(0, time1)]
    distances = [Vox * t for t in np.linspace(0, time)]
    
    plt.plot(xs, ys)
    
    plt.plot(distances, heights)
    
    plt.show()
    
```
